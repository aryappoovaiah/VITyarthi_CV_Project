```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="PlantDoc-4/data.yaml",
    epochs=30,
    imgsz=640,
    batch=16
)
```

```python
metrics = model.val()
print(metrics)
```

```python
results = model.predict(
    source="PlantDoc-4/test/images",
    conf=0.25,
    save=True
)
```
